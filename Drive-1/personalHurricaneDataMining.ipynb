```python
import pandas as pd
import os
import numpy as np
import requests
import json
import sys
import smtplib
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
```

```python


today = datetime.today()
three_months_ago = today - relativedelta(months=3)
date_range = {
    "begin_date": three_months_ago.strftime("%Y%m%d"),
    "end_date": today.strftime("%Y%m%d")
}
```

```python
url = "https://api.tidesandcurrents.noaa.gov/api/prod/datagetter"
params = {"begin_date" : date_range["begin_date"],
          "end_date" : date_range["end_date"],
          "station" : "8724580",
          "product" : "wind",
          "time_zone" : "lst_ldt",
          "interval" : "h",
          "units" : "english",
          "format": "json",
          "latitude": "33.7825",
          "longitude": "-117.2286",
          }
```

```python
threshold = 15.0
```

```python
response = requests.get(url, params=params)
data = response.json()
df = pd.DataFrame(data["data"])
df['timestamp'] = pd.to_datetime(df['t'])
df.set_index('timestamp', inplace=True)
df.head()

```

```python
#check high tide values
for index, row in df.iterrows():
    try:
      water_level = float(row['s'])
    except (ValueError, TypeError):
      continue
    if water_level >= threshold:
      print(f"High tide alert! Water level: "
          f"{water_level} Mean Lower Low Water at {index}")
    if(water_level >= threshold):
     print(f"High tide alert! Water level: "
          f"{water_level} Mean Lower Low Water at {index}")
  #  send_alert
```
